```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS demo
    URL "abfss://demo@stgdeltadatabricks.dfs.core.windows.net/"
    WITH (STORAGE CREDENTIAL `databricks-sc`)
```

```sql
create catalog demo
managed location 'abfss://demo@stgdeltadatabricks.dfs.core.windows.net/'
```

```sql
create schema demo.delta_lake
managed location 'abfss://demo@stgdeltadatabricks.dfs.core.windows.net/delta_lake'
```

```sql
drop table if exists demo.delta_lake.companies;
create table if not exists demo.delta_lake.companies
(
  name string,
  founded_date date,
  country string
);
```

```sql
desc extended demo.delta_lake.companies
```
