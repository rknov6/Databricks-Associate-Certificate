```sql
create or replace table demo.delta_lake.silver_companies
( 
  name string,
  founded_date date,
  country string
)
comment 'This table has companies which are successful'
tblproperties('sensitive' = 'true');

insert into demo.delta_lake.silver_companies(name,founded_date,country) values ('Microsoft',"1976-09-04","USA"),('Google',"1998-09-04","USA"),('Alibaba',"1999-09-04","China");
```

```sql
select * from demo.delta_lake.silver_companies
```

```sql
desc history demo.delta_lake.silver_companies
```

```sql
create or replace table demo.delta_lake.gold_companies
( 
  name string,
  founded_date date,
  country string
)
comment 'This table has companies which are successful'
tblproperties('sensitive' = 'true');

insert into demo.delta_lake.gold_companies(name,founded_date,country) values ('Microsoft',"1976-09-04","USA"),('Google',"1998-09-04","USA")
```

```sql
select * from demo.delta_lake.gold_companies
```

```sql
insert overwrite demo.delta_lake.gold_companies
select * from demo.delta_lake.silver_companies;
select * from demo.delta_lake.gold_companies;
```

```sql
create or replace table demo.delta_lake.gold_companies_partitioned
( 
  name string,
  founded_date date,
  country string
)
partitioned by (country);

insert into demo.delta_lake.gold_companies_partitioned values ('Microsoft',"1976-09-04","USA"),('Google',"1998-09-04","USA"),('Alibaba',"1999-09-04","China");
```

```sql
select * from demo.delta_lake.gold_companies_partitioned;
```
