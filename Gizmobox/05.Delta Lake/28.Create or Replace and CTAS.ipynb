```sql
drop table if exists demo.delta_lake.companies;
create table if not exists demo.delta_lake.companies
( 
  company_id bigint generated always as identity(start with 1 increment by 2),
  name string,
  founded_date date not null comment 'Founded Date',
  country string
)
comment 'This table has companies which are successful'
tblproperties('sensitive' = 'true')
```

```sql
insert into demo.delta_lake.companies(name,founded_date,country) values ('Microsoft',"1976-09-04","USA"),('Google',"1998-09-04","USA"),('Alibaba',"1999-09-04","China")
```

```sql
create or replace table demo.delta_lake.companies
( 
  company_id bigint generated always as identity(start with 1 increment by 2),
  name string,
  founded_date date not null comment 'Founded Date',
  country string
)
comment 'This table has companies which are successful'
tblproperties('sensitive' = 'true');

insert into demo.delta_lake.companies(name,founded_date,country) values ('Microsoft',"1976-09-04","USA"),('Google',"1998-09-04","USA"),('Alibaba',"1999-09-04","China");
```

```sql
create or replace table demo.delta_lake.companies_china
as
select * from demo.delta_lake.companies where country = 'China';
select * from demo.delta_lake.companies_china;
```
