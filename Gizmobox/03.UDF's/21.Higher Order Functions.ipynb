```sql
create or replace temp view order_items
as
select * from 
values (1,array("mobiles","TV","Headphones"))
       ,(2,array("laptop","Tables","Chairs"))
       ,(3,array("Water","Beach","Mountains"))
as orders(order_id,items)
```

```sql
select *,explode(items) from order_items
```

```sql
select *,
explode(TRANSFORM(items,x -> upper(x))) as upper_items
from order_items
```

```sql
select 
   order_id,
   FILTER(items, x -> x like "%s")
   from order_items
```

```sql
select 
   order_id,
   EXISTS(items,x -> x = "mobiles") as existsfun
from order_items
```

```sql
create or replace temp view order_items
as
select *
 from 
    values
      (1,array(
            named_struct("name","smartphone","price",699),
            named_struct("name","laptop","price",1199),
            named_struct("name","monitor","price",399)
      )),
    (2,array(
         named_struct("name","tablet","price",599),
         named_struct("name","smartwatch","price",399),
         named_struct("name","headphones","price",199)
    ))
    ,(3,array(
         named_struct("name","keyboard","price",89),
         named_struct("name","mouse","price",59)
    ))
    as orders(order_id,items)
```

```sql
with cte as (select order_id,explode(items) as abc from order_items)
select order_id,abc.name,abc.price from cte
```

```sql
select Order_id,
TRANSFORM(items,x -> named_struct("name",upper(x.name),"price",x.price * 1.1)) as new_items
from order_items
```

```sql
select Order_id,
AGGREGATE(items,0,(ac,x) -> ac + x.price) as new_items
from order_items
```

#### Map Functions

```sql
create or replace temp view order_items_price
as
select * from values
    (1,map("smartphone",699,"laptop",1199,"monitor",299)),
    (2,map("tablet",599,"headphone",199,"smartwatch",299)),
    (3,map("keyboard",89,"mouse",59))
  as orders(order_id,items)
```

```sql
select *,explode(items) from order_items_price
```

```sql
select
order_id,
TRANSFORM_KEYS(items, (item1,price) -> upper(item1)) as upper_items
from order_items_price
```

```sql
select
order_id,
TRANSFORM_VALUES(items, (item1,price) -> price * 0.1) as upper_items
from order_items_price
```

```sql
select 
   order_id,
   MAP_FILTER(items,(item,price) -> price > 500 and item = "smartphone") as filteritems
   from order_items_price
```
