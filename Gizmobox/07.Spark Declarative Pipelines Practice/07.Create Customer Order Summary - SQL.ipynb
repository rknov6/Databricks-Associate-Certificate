```sql
create or refresh materialized view customer_order_summary
comment "this is gold table"
tblproperties ("quality" = "gold")
as
select
c.customer_id,
c.customer_name,
c.date_of_birth,
c.telephone,
c.email,
a.address_line_1,
a.city,
a.state,
a.postcode,
count(distinct o.order_id) as total_orders,
sum(o.quantity) as total_items_ordered,
sum(o.quantity * o.price) as total_amount_ordered
from live.silver_customers c join live.silver_addresses a on c.customer_id = a.customer_id
join live.silver_orders_clean o on o.customer_id = c.customer_id
where a.__END_AT is null
group by all
```
