```python
from pyspark.sql.functions import col,max
```

```python
df = spark.table("gizmobox.bronze.py_addresses")
df.display()
```

```python
df_update = (
            df.groupBy(col("customer_id"))
              .pivot("address_type",["billing","shipping"])
              .agg(
                  max("address_line_1").alias("address_line_1")
                  ,max("city").alias("city")
                  ,max("state").alias("state")
                  ,max("postcode").alias("postcode")
              )
            )                                                                         
```

```python
df_update.writeTo("gizmobox.silver.py_addresses").createOrReplace()
```
