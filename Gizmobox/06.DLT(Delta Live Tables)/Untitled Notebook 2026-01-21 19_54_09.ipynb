```python
from pyspark.sql.functions import explode,col
```

```python
df = spark.read.json("/Volumes/circuitbox/landing/operationaldata/orders/orders_day2.json")
df2 = df.withColumn("new_items",explode(col("items")))
df2.select("*",col("new_items.category"),col("new_items.item_id").alias("iem_id"),col("new_items.name").alias("name"),col("new_items.price").alias("price"),col("new_items.quantity").alias("quantity")).drop(col("items"),col("new_items")).display()
```
