```python
from pyspark.sql.functions import col,when,date_format
```

```python
df = spark.table("gizmobox.bronze.py_payments")
```

```python
df_status = (
             df.withColumn("payment_status",when(col("payment_status") == 1,"Success").when(col("payment_status") == 2,"Pending").when(col("payment_status") == 3,"Failed").otherwise("Cancelled"))
               .withColumn("payment_date",date_format(col("payment_timestamp"),"yyyy-MM-dd").cast("date"))
               .withColumn("payment_time",date_format(col("payment_timestamp"),"HH:mm:ss"))
               .select("payment_id","order_id","payment_status","payment_method","payment_date","payment_time")
            )
```

```python
df_status.writeTo("gizmobox.silver.py_payments").createOrReplace()
```
