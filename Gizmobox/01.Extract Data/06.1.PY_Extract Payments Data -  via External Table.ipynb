```python
from pyspark.sql.types import StructField,StructType,IntegerType,StringType,TimestampType
```

```python
payments_schema = StructType([
                              StructField("payment_id",IntegerType()),
                              StructField("order_id",IntegerType()),
                              StructField("payment_timestamp",TimestampType()),
                              StructField("payment_status",IntegerType()),
                              StructField("payment_method",StringType())
                              ])
```

```python
schema = "payment_id int,order_id int,payment_timestamp timestamp,payment_status int,payment_method string"
```

```python
df = spark.read.schema(payments_schema).csv("abfss://gizmobox@stgdeltadatabricks.dfs.core.windows.net/landing/external/payments/")
```

```python
df.writeTo('gizmobox.bronze.py_payments').createOrReplace()
```
