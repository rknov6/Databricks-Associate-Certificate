```sql
select 
fixed_value.order_id,
array_distinct(fixed_value.items)
 from gizmobox.silver.orders_json
```

```sql
create or replace temp view tv_orders_array
as
select 
fixed_value.customer_id,
fixed_value.order_id,
fixed_value.order_date,
fixed_value.order_status,
fixed_value.payment_method,
fixed_value.total_amount,
fixed_value.transaction_timestamp,
explode(array_distinct(fixed_value.items)) as item
from gizmobox.silver.orders_json;
```

```sql
drop table if exists gizmobox.silver.orders;
create table if not exists gizmobox.silver.orders
as
select
order_id,
customer_id,
order_date,
order_status,
payment_method,
total_amount,
transaction_timestamp,
item.category,
item.item_id,
item.name,
item.price,
item.quantity,
item.details.brand,
item.details.color
from  tv_orders_array
```
