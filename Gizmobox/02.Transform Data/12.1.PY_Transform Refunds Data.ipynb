```python
from pyspark.sql.functions import col,date_format,split
```

```python
df = spark.table("gizmobox.bronze.py_refunds")
df.display()
```

```python
df_update = (
              df.withColumn("refund_date",date_format(col("refund_timestamp"),"yyyy-MM-dd").cast("date"))
                .withColumn("refund_time",date_format(col("refund_timestamp"),"HH:mm:ss"))
                .withColumn("refund_source",split(col("refund_reason"),":").getItem(1))
                .withColumn("refund_reason",split(col("refund_reason"),":").getItem(0))
                .drop(col("refund_timestamp"))
)
```

```python
df_update.writeTo("gizmobox.silver.py_refunds").createOrReplace()
```
