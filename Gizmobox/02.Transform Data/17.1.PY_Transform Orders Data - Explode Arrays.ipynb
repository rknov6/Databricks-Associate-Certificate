```python
from pyspark.sql.functions import col,explode,explode_outer,array_distinct
```

```python
df = spark.table("gizmobox.silver.py_orders_json")
```

```python
df_explode = df.select(col("fixed_value.customer_id"),"fixed_value.order_id","fixed_value.order_date","fixed_value.order_status",
"fixed_value.payment_method","fixed_value.total_amount","fixed_value.transaction_timestamp",explode(array_distinct("fixed_value.items")).alias("items"))
```

```python
df_final = df_explode.select(col("customer_id"),"order_id","order_date","order_status",
"payment_method","total_amount","transaction_timestamp","items.category","items.item_id","items.name","items.price","items.quantity","items.details.brand","items.details.color")
```

```python
df_final.writeTo("gizmobox.silver.py_orders").createOrReplace()
```
