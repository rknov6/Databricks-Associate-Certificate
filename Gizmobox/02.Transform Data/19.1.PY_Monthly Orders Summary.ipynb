```python
from pyspark.sql.functions import col,date_format,sum,count,count_distinct
```

```python
df = spark.table("gizmobox.silver.py_orders")
```

```python
df_agg = df.groupBy(col("customer_id"),date_format(col("order_date"),"yyyy-MM").alias("order_month")) \
           .agg(sum(col("quantity")).alias("total_items_bought"),count_distinct(col("order_id")).alias("total_orders"),sum(col("price") * col("quantity")).alias("total_amount"))
```

```python
df_agg.writeTo("gizmobox.gold.py_order_summary_monthly").createOrReplace()
```
