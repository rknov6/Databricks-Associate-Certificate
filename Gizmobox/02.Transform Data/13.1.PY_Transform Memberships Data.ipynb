```python
from pyspark.sql.functions import col,regexp_extract
```

```python
df = spark.table("gizmobox.bronze.py_memberships")
```

```python
df_update = df.withColumn("customer_id",regexp_extract(col("path"),'.*/([0-9]+)\\.png$',1)) \
              .withColumnRenamed("content","membership_card") \
              .select("customer_id","membership_card")
```

```python
df_update.writeTo("gizmobox.silver.py_memberships").createOrReplace()
```

```python
%sql
select * from gizmobox.silver.py_memberships;
```
