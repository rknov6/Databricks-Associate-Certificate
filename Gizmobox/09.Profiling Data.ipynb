#### Using DBUTILS

```sql
%python
df = spark.table("gizmobox.bronze.v_customers")
dbutils.data.summarize(df)
```

```sql
select count(*),count(customer_id)
from gizmobox.bronze.v_customers;
```

```sql
select count_if(customer_id is null) as custid,count_if(email is null) as emailid,count_if(telephone is null)
from gizmobox.bronze.v_customers
```

```sql
select count(*)
from gizmobox.bronze.v_customers
where customer_id is null
```

```sql
select count(customer_id),count(distinct customer_id)
from gizmobox.bronze.v_customers
```

```sql
select customer_id,count(*)
from gizmobox.bronze.v_customers
group by customer_id
having count(*) > 1
order by 2 desc
```
