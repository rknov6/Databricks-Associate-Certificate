#### Query Folder

```python
df = spark.read.json("/Volumes/gizmobox/landing/operational/customers/")
df.display()
```

```python
df2 = df.agg(count(col("customer_id")).alias("cnt"),countDistinct(col("customer_id")).alias("distinct count"))
df2.display()
```

#### Add metadata column to existing dataframe

```python
df2 = df.select(col("_metadata.file_path"),"*")
df2.display()
```

#### Create temporary view

```python
df.createOrReplaceTempView("tv_customers")
```

```python
spark.table("tv_customers").display()
```

```python
spark.sql("""
          select * from tv_customers
          """).display()
```

#### Create global temporary view

```python
df.createOrReplaceGlobalTempView("global_tv_customers")
```

```python
spark.table("global_temp.global_tv_customers").display()
```

#### Create bronze table for customers data

```python
df2.write.mode("overwrite").format("delta").saveAsTable("gizmobox.bronze.py_customers")
```

```python
df2.writeTo("gizmobox.bronze.py_customers").createOrReplace()
```

```python
spark.table("gizmobox.bronze.py_customers").display()
```
